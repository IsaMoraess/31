# Construindo Plots

```python
# O matplotlib.pyplot e uma coleção de funções e estilos do Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# O metodo plot() define os eixos do grafico
plt.plot([1, 3, 5],[2, 4, 7])
plt.show()
```

```python
x = [2, 3, 5]
y = [3, 5, 7]
```

```python
plt.plot(x, y)
plt.xlabel( 'Variavel 1')
plt.ylabel( 'Variavel 2 ')
plt.title( 'Teste Plot')
plt.show()
```

```python
x2 = [1, 2, 3]
y3 = [11, 12, 15]
```

```python
plt.plot(x2, y3, label = 'Grafico com Matplotlib')
plt.legend()
plt.show()
```
